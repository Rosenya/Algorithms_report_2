## Cel 
Celem zadania jest zaprojektowanie oraz implementacja programu w języku C++, który symuluje system obsługi zgłoszeń użytkowników (Ticket System) z wykorzystaniem różnych struktur danych.

## Zadanie 
Napisz program w języku C++, który symuluje system obsługi zgłoszeń użytkowników.
Program powinien wykorzystywać różne struktury danych do zarządzania zgłoszeniami.

Każde zgłoszenie powinno być strukturą wg wzoru poniżej:

```python
struct Ticket
{
    int id;
    string name;
    int priority;
};
```

Program musi wykorzystywać:

* listę jednokierunkową
* listę dwukierunkową
* stos
* kolejkę
* tablicę mieszającą
* kolejkę priorytetową (korzystającą ze struktury kopca)

Lista jednokierunkowa — przechowywanie wszystkich zgłoszeń

Operacje:
- dodaj zgłoszenie
- usuń zgłoszenie
- wyświetl listę

Lista dwukierunkowa — zapamiętywanie operacji użytkownika

Operacje:
- zapis historii
- przechodzenie w przód i w tył

Np.:

```python
Dodano zgłoszenie
Usunięto zgłoszenie
Zmieniono priorytet
```

Kolejka — obsługa zgłoszeń FIFO

Operacje:
- enqueue
- dequeue

Stos — cofanie ostatnich operacji (nie musi wykonywać tej akcji, ale musi zaktualizować stos)

Operacje:
- push operacji
- undo

Tablica mieszająca — szybkie wyszukiwanie po ID

Operacje:
- insert
- search
- remove

Kolejka priorytetowa — obsługa zgłoszeń wg priorytetu

Operacje:
- push
- pop


## Problem!
Jak połączyć obsługiwanie według priorytetu wraz z kolejnością dodania do kolejki?
Kiedy wproadzono kilka zgłoszeń o takim samym priorytecie - należy je obsłużyć według starszeństwa, czyli ten który był pierwszy wprowadzony, powinien być pierwszy obsłużony.

## Interface
Przykładowy wygląd menu głównego:

1 - Pozwala użytkownikowi wprowadzić zgłoszenie 
2 - Pokazuje listę aktualnych zgłoszeń
3 - Znajduje zgłoszenie po nazwie lub ID
4 - Wypisuje informację o tym jakie zgłoszenie jest obsługiwane a następnie jest usuwane z kolejki priorytetowej
5 - Wyświetla historię operacji
6 - Wyświetla komunikat jaka operacja jest cofana a następnie jest pokazywana aktualna historia. (OStatnia operacja nie musi być wykonywana przez program)
0 - Zakończenie programu
